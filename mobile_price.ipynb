<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Librarys
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Import required libraries : 
</font>
</p>

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Read data
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    From mobile_prices file read dataset :
</font>
</p>

```python
df = pd.read_csv('mobile_prices.csv')
df.head()
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Labels and Features
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Get features as X and labels as y<br>
    X : Pregnancies to Age<br>
    y : contain outcome
</font>
</p>

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Normalize
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    We equate all data between 0 and 1 by scaler from sklearn :
</font>
</p>

```python
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X=X)

X
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Train and Test
</font>
</h2>
<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Split of train and test data with train_test_split from sklearn
</font>
</p>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train , y_test = train_test_split(X, y, test_size=0.2)
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Naive bayes
</font>
</h2>

```python
from sklearn.naive_bayes import GaussianNB

naive_model = GaussianNB()
naive_model.fit(X=X_train,y=y_train)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Test model : 
</font>
</p>

```python
y_pred_test_naive = naive_model.predict(X_test)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Calclate accuracy :
</font>
</p>

```python
accuracy_naive = accuracy_score(y_pred=y_pred_test_naive, y_true=y_test)
accuracy_naive
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Decision Tree
</font>
</h2>

```python
from sklearn.tree import DecisionTreeClassifier

model_tree = DecisionTreeClassifier(max_depth=32, min_samples_split=2, min_samples_leaf=3, max_features=14)

model_tree.fit(X=X_train,y=y_train)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Test model : 
</font>
</p>

```python
y_pred_test_tree = model_tree.predict(X_test)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Calclate accuracy :
</font>
</p>

```python
accuracy_tree = accuracy_score(y_pred=y_pred_test_tree, y_true=y_test)
accuracy_tree
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Random Forest
</font>
</h2>

```python
from sklearn.ensemble import RandomForestClassifier

model_forest = RandomForestClassifier(n_estimators=100, max_depth=32, min_samples_split=2, min_samples_leaf=2, max_features=16)

model_forest.fit(X=X_train, y=y_train)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Test model : 
</font>
</p>

```python
y_pred_test_forest = model_forest.predict(X_test)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Calclate accuracy :
</font>
</p>

```python
accuracy_forest = accuracy_score(y_pred=y_pred_test_forest, y_true=y_test)
accuracy_forest
```

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
Support Vector Machian (SVM)
</font>
</h2>

```python
from sklearn import svm

model_svm = svm.SVC(kernel='linear')

model_svm.fit(X=X_train,y=y_train)
```

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Test model : 
</font>
</p>

<p dir=rtl style="direction: ltr ;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    Calclate accuracy :
</font>
</p>

<h2 align=left style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
k nearest neighbors (KNN)
</font>
</h2>
